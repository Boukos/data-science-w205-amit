## connecting via ssh to an EC2 instance on Amazon

### launch an instance

steps you need to take 

1. get an AWS (Amazon Web Services) account
1. go to EC2 (Elastic Cloud Compute) Dashboard
  - click >> **Launch Instance**
1. find an AMI (Amazon Machine Image)
  - try this one >> **UCB W205 Spring 2016 - ami-be0d5fd4**
1. pick an instance type
  - **t1.micro** is fine for just getting started
  - you will need **m3.medium** or **m3.large** for later labs
1. configure instance details 
  - pick an availability zone that you like/remember
  - everything you do should be in the same zone
1. add storage
  - skip for now
  - you can specify additional "hard drives" that you would like your VM (virtual machine) to access
1. add tags 
  - skip
  - conveniently name and tag stuff for when you have 100's of VM in your account and you can't find any of them
1. configure security groups
  - this is **SUPER IMPORTANT**
  - you must have port 22 open for ssh otherwise you can't login
  - you can specify open other ports now 
  - you can change this later too 
  - an open port is a "window" into your running VM
  - any open port on you VM is a security risk so you want to open only the ports you need
  - for example, if you want to run a webserver you would want port 80 open to http requests
  - for example, if you want to run a secure webserver you would want port 443 open to https requests
1. review and launch  
  - if everything seems ok, then launch
1. select key pair
  - a popup screen will appear
  - this is also **SUPER IMPORTANT**
  - without the proper key pair you can't login into your new instance
  - create is new one if you haven't done this before
  - or choose an existing one so you don't end up w/ too many
  - i called mine "foo" which allows me to download a file called **`foo.pem`**
  - once you download this secret file you can LAUNCH
1. go to the EC2 dashboard
  - go to "running instances" to check on the status
  - it takes a few minutes for the instance to get created and boot up
  

### connect to your instance

steps you need to take 

1. find your instance on the EC2 dashboard
  - go to **running instances** to check on the status
  - click on the line with the instance you just created and you the description should populate below
  
2. figure out the IP address or hostname of your new instance
  - look at **Public DNS** to get the fully qualified hostname
  - look at **Public IP** to get the IP address
  
3. get yourself a command line
  - on Windows startup **gitbash** or something
  - on Windows using **Putty** - try not to go down this route
  - on Mac start the **terminal** app
  - from any machine, click on Actions > Connect > **Java SSH client**
    - you might have to mess w/ some Java security settings on your computer
    
4. run this command to ssh to your machine as root
  - you have to be in the same directory where you saved the **foo.pem** file
  - in general it is bad practice to login as root
  - however, all the instructions in all the labs have you do it anyways!
  ```
  ssh -i foo.pem root@ec2-184-73-20-249.compute-1.amazonaws.com
  ```
  
  you can also run this command 
  ```
  ssh -i foo.pem root@184.73.20.249
  ```
  
  you will probably see this error
  ```
@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
@         WARNING: UNPROTECTED PRIVATE KEY FILE!          @
@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
Permissions 0644 for 'foo.pem' are too open.
It is required that your private key files are NOT accessible by others.
This private key will be ignored.
bad permissions: ignore key: foo.pem
```

5. change permissions of the pem file
  - switch directory to where you saved the pem file
  - run this command to make tweak permissions (only read/write access by you)
  ```
  chmod 400 foo.pem
  ```
  - now try ssh command again

6. success

  you should see something like this if you are able to login
  
  ```
  Welcome to a virtual machine image brought to you by RightScale!

  ```

### mess around with your instance

here are some basic commands to try

1. who are you logged in as?
```
whoami
```

1. what is the current directory?
```
pwd
```

1. list directory contents
```
ls
```

1. show filesystem (-h is for human readable output)
```
df -h
```

1. check out the root filesystem
```
ls /
```

1. get more details on the root filesystem
```
ls -la /
```

1. check the location of the python executable
```
which python
```

1. check to see what version of python is in the system directory
```
python --verison
```


### terminate your instance

1. you can either stop or terminate an EC2 instance
  - stop means "keep it alive, just turn the power off so i can restart later"
  - terminate means "throw the computer away forever"
  
1. how you might choose
  - stop when you want to preserve your work but don't want to pay hourly compute fees
  - you still have to pay the storage charges (which are minimal)
  - terminate when you have no need for this VM anymore
  - or if you don't even want to pay the storage charges which can get high for machines w/ large hard drives
  
1. **important:** if you dont stop the instance you can rack up a big AWS bill
